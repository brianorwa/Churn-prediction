# Data Cleaning

```python
class DataChecker:
    def __init__(self, df):
        self.df = df

    def check_missing_values(self):
        missing_values = self.df.isnull().sum()
        if missing_values.sum() == 0:
            print("No missing values found.")
        else:
            print("Missing values per column:")
            print(missing_values)

    def check_duplicates(self):
        duplicates = self.df.duplicated()
        if duplicates.sum() == 0:
            print("No duplicates found.")
        else:
            print("Number of duplicate rows:", duplicates.sum())

    def detect_outliers(self, columns=None, threshold=1.5):
        if columns is None:
            columns = self.df.select_dtypes(include='number').columns

        outlier_indices = []

        for column in columns:
            # Calculate IQR
            Q1 = self.df[column].quantile(0.25)
            Q3 = self.df[column].quantile(0.75)
            IQR = Q3 - Q1

            # Define upper and lower bounds for outliers
            lower_bound = Q1 - threshold * IQR
            upper_bound = Q3 + threshold * IQR

            # Identify potential outliers
            outliers = self.df[(self.df[column] < lower_bound) | (self.df[column] > upper_bound)]
            outlier_indices.extend(outliers.index)

        # Remove duplicate indices
        outlier_indices = list(set(outlier_indices))

        if len(outlier_indices) == 0:
            print("No outliers found.")
        else:
            print("Potential outliers indices:", outlier_indices)

data_checker = DataChecker(df)
data_checker.check_missing_values()
data_checker.check_duplicates()
data_checker.detect_outliers(columns=['account length', 'total day minutes', 'total night minutes'])
```

```python
class CategoricalValueCounter:
    def __init__(self, df, categorical_columns):
        self.df = df
        self.categorical_columns = categorical_columns

    def value_counts(self):
        value_counts_dict = {}

        for column in self.categorical_columns:
            value_counts_dict[column] = self.df[column].value_counts().to_dict()

        return value_counts_dict
categorical_columns = ['state', 'international plan', 'voice mail plan', 'churn']
value_counter = CategoricalValueCounter(df, categorical_columns)
result_value_counts = value_counter.value_counts()

# Display the value counts
for column, counts in result_value_counts.items():
    print(f"Value counts for {column}:")
    print(counts)
    print()
```

```python
class CategoricalValueCounter:
    def __init__(self, df, categorical_columns):
        self.df = df
        self.categorical_columns = categorical_columns

    def value_counts_percentage(self):
        percentages_dict = {}

        for column in self.categorical_columns:
            value_counts = self.df[column].value_counts()
            percentages = value_counts / len(self.df) * 100
            percentages_dict[column] = percentages.to_dict()

        return percentages_dict
categorical_columns = ['state', 'international plan', 'voice mail plan', 'churn']
value_counter = CategoricalValueCounter(df, categorical_columns)
result_value_percentages = value_counter.value_counts_percentage()

# Display the value counts in percentage form
for column, percentages in result_value_percentages.items():
    print(f"Value percentages for {column}:")
    print(percentages)
    print()
```

# 1. Checking for corellations

```python
# Checking for corellated columns
def check_correlations(df, threshold=0.95, print_info=True):
   
    # Select numeric columns
    numeric_columns = df.select_dtypes(include='number').columns
    
    # Calculate correlation matrix
    correlation_matrix = df[numeric_columns].corr()

    # Identify highly correlated pairs
    correlated_pairs = []
    for i in range(len(correlation_matrix.columns)):
        for j in range(i+1, len(correlation_matrix.columns)):
            if abs(correlation_matrix.iloc[i, j]) >= threshold:
                correlated_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j]))

    # Print or return information about correlated features
    if print_info:
        if not correlated_pairs:
            print("No highly correlated features found.")
        else:
            print("Highly Correlated Feature Pairs:")
            for pair in correlated_pairs:
                print(f"{pair[0]} and {pair[1]} are highly correlated.")
    else:
        return correlated_pairs

check_correlations(df)
```

Most of the features are not correlated however some do share a perfect correlation.

- Total day charge and total day minutes features are fully positively correlated.
- Total eve charge and total eve minutes features are fully positively correlated.
- Total night charge and total night minutes features are fully positively correlated.
- Total int charge and total int minutes features are fully positively correlated.

It makes sense for these features to be perfectly correlated because the charge is a direct result of the minutes used.

```python
#Dropping one of each correlated features
def drop_correlated_features(df, threshold=0.95, drop_method='first', print_info=True):
   
    # Select numeric columns
    numeric_columns = df.select_dtypes(include='number').columns
    
    # Calculate correlation matrix
    correlation_matrix = df[numeric_columns].corr()

    # Identify highly correlated pairs
    correlated_pairs = []
    for i in range(len(correlation_matrix.columns)):
        for j in range(i+1, len(correlation_matrix.columns)):
            if abs(correlation_matrix.iloc[i, j]) >= threshold:
                correlated_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j]))

    # Drop one feature from each correlated pair
    df_dropped = df.copy()
    for pair in correlated_pairs:
        if drop_method == 'first':
            df_dropped = df_dropped.drop(pair[1], axis=1)
        elif drop_method == 'last':
            df_dropped = df_dropped.drop(pair[0], axis=1)
        elif drop_method == 'random':
            # Drop a random feature from the pair
            df_dropped = df_dropped.drop(np.random.choice(pair), axis=1)

    # Print or return information about dropped features
    if print_info:
        if not correlated_pairs:
            print("No highly correlated features found.")
        else:
            print(f"Dropped one feature from each correlated pair using drop_method='{drop_method}'.")
            for pair in correlated_pairs:
                print(f"Dropped: {pair[1]}")

    return df_dropped

# Call the function
df0 = drop_correlated_features(df, drop_method='first')
```
